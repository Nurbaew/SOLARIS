```python
import datetime
import calendar
from dateutil.relativedelta import relativedelta
import pandas as pd
```

```python
start = int(input())
end = int(input())
is_window = int(input())
schedule = []
for year in range(start,end+1): # для каждого года от sart до end 
    for month in range(1,13):               #для каждого месяца года из цикла
        oos_start = datetime.datetime(year,month,1)     #определетс первый день месяца
        oos_end = datetime.datetime(year,month,calendar.monthrange(year,month)[1])   #определяется последни день месяца (размах месяца + последний день)
        #print(oos_start,oos_end)
        is_start = oos_start - relativedelta(months=is_window)
        is_end = oos_start - relativedelta(days=1)
        schedule.append({
            'OOS_START':oos_start.strftime('%d.%m.%Y'),
            'OOS_END':oos_end.strftime('%d.%m.%Y'),
            'IS_START':is_start.strftime('%d.%m.%Y'),
            'IS_END':is_end.strftime('%d.%m.%Y'),
        })
df = pd.DataFrame(schedule)
df.index = range(1, len(df) + 1)
df.to_csv('CSV/schedule.CSV', sep = ';',index=False)
print(df.head(3))
```
