```python
import pandas as pd
import yfinance as yf

#загрузка названия тикеров из собсвтенного эксель соубора 
snp500 = pd.read_excel('SP500_list.xlsx')
equity_series = pd.Series(snp500['Ticker'].dropna().str.strip().unique(), name='EQUITY')
#print(equity_series)

#создание пайтон списка
equity_list =['']
crypto_list = ['BTC-USD', 'ETH-USD', 'SOL-USD']
forex_list = ['EURUSD=X', 'GBPUSD=X', 'USDJPY=X', 'AUDUSD=X', 'USDCAD=X']
metals_list = ['GC=F', 'SI=F']  # Золото и Серебро

#создание пандас сериеса
crypto_series = pd.Series(crypto_list, name ='CRYPTO')
forex_series = pd.Series(forex_list,name = 'FOREX')
metals_series = pd.Series(metals_list, name = 'COMODITIES')

fina_pd = pd.concat([equity_series,crypto_series,forex_series,metals_series], axis=1)
fina_pd = fina_pd.fillna('')
fina_pd.to_csv('CSV/TICKERS.csv',index=False)
print(fina_pd)
```
